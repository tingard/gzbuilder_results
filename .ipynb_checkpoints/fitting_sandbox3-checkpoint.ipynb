```python
def rotmx(a):
    return np.array(((np.cos(a), np.sin(a)), (-np.sin(a), np.cos(a))))

def log_spiral(t_a, t_b, A, phi, q=1, roll=0, delta_roll=0, mux=0, muy=0, N=200, **kwargs):
    theta = np.linspace(t_a, t_b, N)
    qmx = np.array(((q, 0), (0, 1)))
    mx = np.dot(rotmx(-roll), np.dot(qmx, rotmx(delta_roll)))
    return ((
        A * np.exp(theta * np.tan(np.deg2rad(phi))) \
        * np.dot(mx, np.array((np.cos(theta), np.sin(theta))))
    ).T + np.array([mux, muy])).T
```
