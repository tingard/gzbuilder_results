```python
import os
import json
import numpy as np
import pandas as pd
from matplotlib import rc
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
gzb_subjects = pd.read_csv('lib/galaxy-builder-subjects.csv', index_col=0)
hart_table = pd.read_csv('../source_files/Hart2017/psi_table.csv', index_col=0)
```

```python
nsa_cat = fits.open('../source_files/NSA_GalaxyZoo.fits')
```

```python
main_run_mask = np.isin(gzb_subjects.subject_set_id, (20561, 20751, 21182))
subject_sets = gzb_subjects.subject_set_id[main_run_mask].replace({20561: 'Subset 1', 20751: 'Subset 2', 21182: 'Validation Set'})
```

```python
gzb_dr7ids = gzb_subjects[main_run_mask].metadata.apply(lambda a: pd.Series(json.loads(a)))['SDSS dr7 id'].dropna().astype(np.int64)
```

```python
lookup = {np.int64(v): k for k, v in gzb_dr7ids.iteritems()}
```

```python
nsa_data = nsa_cat[1].data
nsa_data = nsa_data[np.isin(nsa_data['DR7OBJID'], hart_table.index.values)]
```

```python
subject_set_by_id = pd.concat((gzb_dr7ids, subject_sets), axis=1).set_index('SDSS dr7 id')
```

```python
mass_df = pd.concat((
    hart_table,
    pd.Series(nsa_data['z'].byteswap().newbyteorder(), nsa_data['DR7OBJID'].byteswap().newbyteorder(), name='Z'),
), axis=1).query('Z > 0')

f, ax = plt.subplots(ncols=2, figsize=(12, 4), dpi=100, gridspec_kw=dict(width_ratios=(7,1)), sharey=True)
plt.sca(ax[0])
mass_df.plot.scatter('Z', 'logMstar', ax=plt.gca(), s=1, c='k', alpha=0.7, logy=False, label='Entire stellar mass-complete sample')
mass_df.reindex(gzb_dr7ids.values).plot.scatter('Z', 'logMstar', ax=plt.gca(), s=10, c='C3', alpha=1, logy=False, label='Galaxy Builder\ngalaxies')
plt.xlabel('Redshift')
plt.ylabel(r'$\log\,\left(M_\ast\,/\,M_\odot\right)$')
plt.legend()
plt.sca(ax[1])
sns.distplot(mass_df.logMstar.dropna(), kde=False, norm_hist=True, bins='scott', vertical=True, color='k', label='')
sns.distplot(mass_df.reindex(gzb_dr7ids.values).logMstar.dropna(), kde=False, norm_hist=True, bins='scott', vertical=True, color='C3', label='')
plt.xlabel('Density')
plt.tight_layout()

plt.savefig('method-paper-plots/stellar_mass_selection_plot.pdf', bbox_inches='tight')
plt.savefig('method-paper-plots/stellar_mass_selection_plot.png', bbox_inches='tight');
```

```python
f, ax = plt.subplots(ncols=1, figsize=(10, 4), dpi=100)
mass_df.plot.scatter('Z', 'logMstar', ax=plt.gca(), s=1, c='k', alpha=0.7, logy=False)
# mass_df.reindex(gzb_dr7ids.values).plot.scatter('Z', 'logMstar', ax=plt.gca(), s=10, c='C3', alpha=1, logy=False)
for i, (idx, group) in enumerate(subject_set_by_id.groupby('subject_set_id')):
    if idx == 'Validation Set':
        kw = dict(s=2, c='C3', zorder=3, alpha=1)
        continue
    else:
        kw = dict(s=15, c='C{}'.format(i), zorder=1, alpha=1)
    mass_df.reindex(group.index).plot.scatter('Z', 'logMstar', ax=plt.gca(), logy=False, label=idx, **kw)
plt.xlabel('Redshift')
plt.ylabel(r'$\log\,\left(M_\ast\,/\,M_\odot\right)$');

# plt.savefig('method-paper-plots/stellar_mass_selection_plot.pdf', bbox_inches='tight')
# plt.savefig('method-paper-plots/stellar_mass_selection_plot.png', bbox_inches='tight')
```
